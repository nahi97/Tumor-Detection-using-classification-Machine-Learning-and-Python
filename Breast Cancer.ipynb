```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
dataset=pd.DataFrame(data=data['data'],columns=data['feature_names'])
dataset
```

```python
dataset.info()
```

```python
# Correlation Plot
corr = dataset.corr()
corr
```

```python
# making a heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14, 14))
sns.heatmap(corr)
```

```python
from sklearn.model_selection import train_test_split
X=dataset.copy()
y=data["target"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
```

```python
clf.get_params()
```

```python
X_test
```

```python
predictions=clf.predict(X_test)
predictions
```

```python
#shows difference between tree that doesn't have stopping criteria and the one that does
clf.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
dec_acc= accuracy_score(y_test,predictions)
dec_acc
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions,labels=[0,1])
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,predictions)
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test,predictions,target_names=['Malignant','Benign']))
```

```python
# apply Random Forest Classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
y_pred

rfc_acc=accuracy_score(y_test, y_pred)
rfc_acc
```

```python
results = pd.DataFrame(columns=['Algorithm', 'Accuracy'])
tempResults = pd.DataFrame({'Algorithm': ['Random Forest Classifier Method'], 
							'Accuracy': [rfc_acc]})

results = pd.concat([results, tempResults])
results = results[['Algorithm', 'Accuracy']]
results
```

```python
# apply Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)

# implemented our model through logistic regression
y_pred = lr.predict(X_test)
y_pred

# array containing the actual output
y_test
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(y_test, y_pred)
lr_acc
```

```python
tempResults = pd.DataFrame({'Algorithm': ['Linear Regression Method'], 
							'Accuracy': [lr_acc]})

results = pd.concat([results, tempResults])
results = results[['Algorithm', 'Accuracy']]
results
```
